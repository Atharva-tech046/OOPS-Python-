```python
# del - used to delete object properties or object itself 
#eg 

class student:
    def __init__(self,name):
        self.name=name

S1 = student("Atharva")
del S1                #-->This function deletes the S1 record from the class student
print(S1)  #-------> Hence gives error 
```

```python
#PRIVATE ATTRIBUTES 
#in private attributes these attributes cannot be accessed outside of the class 
#these can only be accessed inside the class unlike the other public attributes that we made in previous codes 
#usually used to in case of sensitive information 

#eg 

class Account:
    def __init__(self,acc_no,acc_pass):     #---> since acc pass is a sensitive info we make it private 
        self.acc_no=acc_no                 
        self.__acc_pass=acc_pass #--> we put __ under the attribute 


acc1 = Account(123,33456)

print(acc1.acc_pass)    #----> This will give an error as it we are calling / prinitng a priv attribute 
```
