```python
#OOP (Object oriented proggramming)

#Main objective of OOP - * to reduce redundancy 
#                        * to increase reuseability 


# CLASS - It is a blueprint for creating objects 

#eg code 

class student :        #---------> CLASS 
    name = "Atharva"

S1 = student()    #------------> OBJECT 
print(S1.name)
```

```python
#eg 2 

class car :                #------------> CLASS 
    brand = "Mercedes"
    model = "GLE"
    colour = "White"

C1 = car()                    #----------> OBJECT
print(C1.colour)
print(C1.model)
print(C1.brand)
```

```python
#eg 3 (using multiple costructors)

class student : 
    def __init__(self):           #----> initialization runs even if we dont write it 
        print("Adding new student to Database...")
    
    def __init__(self,name,marks):  #---> This is a parameterized constructor
        self.name = name
        self.marks = marks
        print("Adding new student to the Database....")

S1 = student("Atharva",99)          #------->This is an Object 
print(S1.name,S1.marks)

S2 = student("Arjun",97)
print(S2.name,S2.marks)
```
