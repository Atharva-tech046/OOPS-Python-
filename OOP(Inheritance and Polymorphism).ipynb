```python
# INHERITANCE - When one child class derives properties and methods from the parent class.
# basically it means that one class inheriting the properties and function of another class
#It promotes hierarchical classification and promotes code reuse 

class Car():
    colour = 'black'
    @staticmethod
    def start():
        print("Car started âœ…")

    @staticmethod
    def stop():
        print("Car stopped ðŸ›‘")

class ToyotaCar(Car):
    def __init__(self,name):
        self.name = name

car1 = ToyotaCar("fortuner")
car2 = ToyotaCar("Camry")

print(car1.start())
print(car1.colour)
```

```python
# TYPES OF INHERITANCE 
# 1) Single level inheritance 
# 2) Multi-level inheritance 
# 3) Multiple inheritance
```

```python
# Single Level Inheritance - In this the a child class is derieved from the parent class 

#eg 

class Animal():                    #--> Parent Class 
    def __init__(self,name):
        self.name = name
    
class app (Animal):                 #---> Child class 
    def __init__ (self,colour,name):
        super().__init__(name)        #---> Calling the constructor of the Parent class 
        self.colour = colour
        
A1 = app("Orange","Tiger")

print([A1.colour,A1.name])

```

```python
# Multi-Level Inheritance - In this another child class is derived from the parent class after the first one 


class Student():
    def __init__(self,name):
        self.name = name

class Marks(Student):
    def __init__(self,name,score):
        super().__init__(name)
        self.score = score

class roll(Marks):
    def __init__(self,name,score,UID):
        super().__init__(name,score)
        self.UID = UID

S1 = roll("Atharva",99,201)    #--> OBJECT

print([S1.name,S1.score,S1.UID])
```

```python
# Multiple Inheritance - In this a child class is derived from to parent classes 

class A():                       #---> Parent class 1
    varA = "Welcome to class A"

class B():
    varB = "Welcome to class B"   #----> Parent class 2

class C(A,B):                       #----> Inheriting from both parent classes 1 and 2
    varC = "Welcome to class C"

C1 = C()    #-----> Object 


print(C1.varA)
print(C1.varB)
print(C1.varC) 
```

```python
# SUPER METHOD - it is used while inheriting the attributes/methods/functions from the parent class or any other subordinate class 
# syntax = super().__init__()

#eg code 

class Car():
    def __init__(self,name):
        self.name = name

class details(Car):
    def __init__(self,name,model,UID):
        super().__init__(name)        #----> using super function to inherit the 
        self.model = model 
        self.UID = UID

C1 = details("Mercedes","GLE",2230)

print(C1.model)
print(C1.name)
print(C1.UID)
```
